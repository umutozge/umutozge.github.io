```python
from functools import wraps
def to_execute(result_key):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            iterator = func(*args, **kwargs)
            final_state = list(iterator)[-1]
            return getattrs(final_state, result_key)
        return wrapper
    return decorator
```
